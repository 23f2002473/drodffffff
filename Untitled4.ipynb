```python
import pandas as pd
import numpy as np
import torch
from datasets import Dataset, load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSeq2SeqLM,
    DataCollatorForSeq2Seq,
    get_scheduler
)
from peft import LoraConfig, get_peft_model, TaskType
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split
from tqdm.auto import tqdm
!pip install evaluate
import evaluate
import time
import matplotlib.pyplot as plt
```

```python
import random
random.seed(42)
np.random.seed(42)
torch.manual_seed(42)

# ============================================================================
# 1. LOAD ALPACA DATA
# ============================================================================
print("="*80)
print("LOADING ALPACA DATA")
print("="*80)

# Load Alpaca dataset from Hugging Face
dataset = load_dataset("tatsu-lab/alpaca")

# Convert to pandas for easier manipulation
df = pd.DataFrame(dataset['train'])

# Filter out rows with missing values
df = df.dropna(subset=['instruction', 'output']).reset_index(drop=True)

# Fill empty 'input' fields with empty string
df['input'] = df['input'].fillna('')

print(f"Total samples: {len(df)}")
print(f"\nSample data:")
print(df.head(3))
```

```python
# 2. TRAIN-VAL-TEST SPLIT
# ============================================================================
print("\n" + "="*80)
print("DATA SPLITTING")
print("="*80)

# Split data: 80% train, 10% val, 10% test
train_df, temp_df = train_test_split(df, test_size=0.2, random_state=42)
val_df, test_df = train_test_split(temp_df, test_size=0.5, random_state=42)

print(f"Train set: {len(train_df)} samples ({len(train_df)/len(df)*100:.1f}%)")
print(f"Validation set: {len(val_df)} samples ({len(val_df)/len(df)*100:.1f}%)")
print(f"Test set: {len(test_df)} samples ({len(test_df)/len(df)*100:.1f}%)")
```

```python
# 3. LOAD MODEL & TOKENIZER
# ============================================================================
print("\n" + "="*80)
print("LOADING FLAN-T5-BASE MODEL")
print("="*80)

model_name = "google/flan-t5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Model: {model_name}")
print(f"Using device: {device}")
print(f"Total parameters: {model.num_parameters():,}")
```

```python
# 4. LORA CONFIGURATION
# ============================================================================
print("\n" + "="*80)
print("CONFIGURING LORA")
print("="*80)

lora_config = LoraConfig(
    r=16,  # LoRA rank
    lora_alpha=32,  # Scaling factor
    target_modules=["q", "v", "k", "o", "wi", "wo"],  # Attention + FFN layers
    lora_dropout=0.1,
    bias="none",
    task_type=TaskType.SEQ_2_SEQ_LM
)

model = get_peft_model(model, lora_config)
trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
total_params = sum(p.numel() for p in model.parameters())

print(f"LoRA Configuration:")
print(f"  Rank: {lora_config.r}")
print(f"  Alpha: {lora_config.lora_alpha}")
print(f"  Target modules: {lora_config.target_modules}")
print(f"  Dropout: {lora_config.lora_dropout}")
print(f"\nTrainable parameters: {trainable_params:,} ({trainable_params/total_params*100:.2f}%)")
print(f"Total parameters: {total_params:,}")

model.to(device)
```

```python
# 5. PREPARE ALPACA DATA
# ============================================================================
print("\n" + "="*80)
print("TASK-ADAPTIVE INSTRUCTION FORMATTING")
print("="*80)

max_input_length = 512
max_target_length = 256

def prepare_alpaca_data(df):
    """
    Prepare Alpaca format data: (instruction + input) -> output
    """
    data = []
    for _, row in df.iterrows():
        # Combine instruction and input
        if row['input'].strip():
            input_text = f"{row['instruction']}\n\nInput: {row['input']}"
        else:
            input_text = row['instruction']

        data.append({
            'input_text': input_text,
            'target_text': row['output']
        })
    return data

train_data = prepare_alpaca_data(train_df)
val_data = prepare_alpaca_data(val_df)
test_data = prepare_alpaca_data(test_df)

print("Example formatted input:")
print(train_data[0]['input_text'][:200] + "...")
print("\nExpected output:")
print(train_data[0]['target_text'][:200] + "...")
```

```python
# 6. TOKENIZATION
# ============================================================================
print("\n" + "="*80)
print("TOKENIZING DATA")
print("="*80)

def preprocess_function(examples):
    model_inputs = tokenizer(
        examples['input_text'],
        max_length=max_input_length,
        truncation=True,
        padding=False
    )

    labels = tokenizer(
        text_target=examples['target_text'],
        max_length=max_target_length,
        truncation=True,
        padding=False
    )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

train_dataset = Dataset.from_list(train_data).map(
    preprocess_function,
    batched=True,
    remove_columns=['input_text', 'target_text']
)
val_dataset = Dataset.from_list(val_data).map(
    preprocess_function,
    batched=True,
    remove_columns=['input_text', 'target_text']
)
test_dataset = Dataset.from_list(test_data).map(
    preprocess_function,
    batched=True,
    remove_columns=['input_text', 'target_text']
)

print("Tokenization complete!")
print(f"Train samples: {len(train_dataset)}")
print(f"Validation samples: {len(val_dataset)}")
print(f"Test samples: {len(test_dataset)}")
```

```python
# 7. DATA LOADERS
# ============================================================================

data_collator = DataCollatorForSeq2Seq(
    tokenizer=tokenizer,
    model=model,
    padding=True,
    label_pad_token_id=-100
)

batch_size = 8
train_dataloader = DataLoader(
    train_dataset,
    shuffle=True,
    batch_size=batch_size,
    collate_fn=data_collator
)
val_dataloader = DataLoader(
    val_dataset,
    batch_size=batch_size,
    collate_fn=data_collator
)
test_dataloader = DataLoader(
    test_dataset,
    batch_size=batch_size,
    collate_fn=data_collator
)
```

```python
# 8. EVALUATION METRICS SETUP
# ============================================================================
print("\n" + "="*80)
print("LOADING EVALUATION METRICS")
print("="*80)

bleu_metric = evaluate.load("sacrebleu")
rouge_metric = evaluate.load("rouge")
bertscore_metric = evaluate.load("bertscore")
```

```python
# 9. OPTIMIZER & SCHEDULER
# ============================================================================
print("\n" + "="*80)
print("TRAINING CONFIGURATION")
print("="*80)

optimizer = torch.optim.AdamW(model.parameters(), lr=3e-4, weight_decay=0.01)

num_epochs = 3
num_training_steps = num_epochs * len(train_dataloader)
num_warmup_steps = 500

lr_scheduler = get_scheduler(
    "cosine",
    optimizer=optimizer,
    num_warmup_steps=num_warmup_steps,
    num_training_steps=num_training_steps
)

print(f"Training for {num_epochs} epochs")
print(f"Total training steps: {num_training_steps}")
print(f"Warmup steps: {num_warmup_steps}")
```

```python
# 10. TRAINING LOOP
# ============================================================================
print("\n" + "="*80)
print("STARTING TRAINING")
print("="*80)

start_time = time.time()
best_val_loss = float('inf')
training_losses = []
validation_losses = []

for epoch in range(num_epochs):
    # ========================================
    # TRAINING
    # ========================================
    model.train()
    total_train_loss = 0
    train_steps = 0

    progress_bar = tqdm(train_dataloader, desc=f"Epoch {epoch+1}/{num_epochs}")

    for batch in progress_bar:
        batch = {k: v.to(device) for k, v in batch.items()}

        outputs = model(**batch)
        loss = outputs.loss

        total_train_loss += loss.item()
        train_steps += 1

        loss.backward()

        # Gradient clipping
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)

        optimizer.step()
        lr_scheduler.step()
        optimizer.zero_grad()

        # Update progress bar
        progress_bar.set_postfix({'loss': f'{loss.item():.4f}'})

    avg_train_loss = total_train_loss / train_steps
    training_losses.append(avg_train_loss)
```

```python
    # VALIDATION
    # ========================================
    model.eval()
    total_val_loss = 0
    val_steps = 0

    print(f"\nValidating...")
    with torch.no_grad():
        for batch in tqdm(val_dataloader, desc="Validation"):
            batch = {k: v.to(device) for k, v in batch.items()}
            outputs = model(**batch)
            loss = outputs.loss
            total_val_loss += loss.item()
            val_steps += 1

    avg_val_loss = total_val_loss / val_steps
    validation_losses.append(avg_val_loss)

    print(f"\n{'='*80}")
    print(f"Epoch {epoch+1}/{num_epochs} Complete")
    print(f"{'='*80}")
    print(f"Average Training Loss: {avg_train_loss:.4f}")
    print(f"Average Validation Loss: {avg_val_loss:.4f}")
    print(f"Learning Rate: {lr_scheduler.get_last_lr()[0]:.2e}")

    # Save best model
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        print(f"✓ New best model! Saving...")
        model.save_pretrained("./flan-t5-alpaca-lora-best")
        tokenizer.save_pretrained("./flan-t5-alpaca-lora-best")

    print()

training_time = time.time() - start_time

print("\n" + "="*80)
print("TRAINING COMPLETE")
print("="*80)
print(f"Training completed in {training_time/60:.2f} minutes")
print(f"Best validation loss: {best_val_loss:.4f}")

# Save final model
model.save_pretrained("./flan-t5-alpaca-lora-final")
tokenizer.save_pretrained("./flan-t5-alpaca-lora-final")
```

```python
# 11. EVALUATION ON TEST SET
# ============================================================================
print("\n" + "="*80)
print("EVALUATING ON TEST SET")
print("="*80)

# Load best model
print("Loading best model...")
from peft import PeftModel
base_model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
model = PeftModel.from_pretrained(base_model, "./flan-t5-alpaca-lora-best")
model.to(device)
model.eval()

all_predictions = []
all_references = []

# Evaluate on subset for speed (or full test set if you prefer)
test_sample_size = min(1000, len(test_data))
test_sample = test_data[:test_sample_size]

print(f"Generating predictions on {test_sample_size} test samples...")
eval_start = time.time()

for example in tqdm(test_sample, desc="Generating predictions"):
    inputs = tokenizer(
        example['input_text'],
        return_tensors="pt",
        max_length=max_input_length,
        truncation=True
    ).to(device)

    with torch.no_grad():
        outputs = model.generate(
            **inputs,
            max_length=max_target_length,
            num_beams=4,
            no_repeat_ngram_size=2,
            early_stopping=True
        )

    prediction = tokenizer.decode(outputs[0], skip_special_tokens=True)
    all_predictions.append(prediction.strip())
    all_references.append(example['target_text'].strip())

inference_time = time.time() - eval_start

# Calculate metrics
print("\nCalculating metrics...")
bleu_result = bleu_metric.compute(
    predictions=all_predictions,
    references=[[r] for r in all_references]
)
rouge_result = rouge_metric.compute(
    predictions=all_predictions,
    references=all_references,
    use_stemmer=True
)
exact_match = sum(p == r for p, r in zip(all_predictions, all_references)) / len(all_predictions) * 100

bertscore_result = bertscore_metric.compute(
    predictions=all_predictions,
    references=all_references,
    lang="en",
    model_type="distilbert-base-uncased"
)

print(f"\n{'='*80}")
print(f"TEST SET RESULTS ({test_sample_size} samples)")
print(f"{'='*80}")
print(f"BLEU-4: {bleu_result['score']:.2f}")
print(f"ROUGE-L: {rouge_result['rougeL']:.4f}")
print(f"ROUGE-1: {rouge_result['rouge1']:.4f}")
print(f"ROUGE-2: {rouge_result['rouge2']:.4f}")
print(f"Exact Match: {exact_match:.2f}%")
print(f"BERTScore F1: {np.mean(bertscore_result['f1'])*100:.2f}")
print(f"Inference time: {inference_time:.2f}s")
print(f"Latency per example: {inference_time/len(all_predictions)*1000:.2f}ms")
```

```python
# 12. BENCHMARK COMPARISONS (ZERO-SHOT & FEW-SHOT)
# ============================================================================
print("\n" + "="*80)
print("BENCHMARK: ZERO-SHOT FLAN-T5-BASE")
print("="*80)

# Load pretrained model (no fine-tuning)
base_model_zeroshot = AutoModelForSeq2SeqLM.from_pretrained(model_name)
base_model_zeroshot.to(device)
base_model_zeroshot.eval()

def evaluate_baseline(model_eval, test_data_subset, model_desc):
    """Evaluate baseline model"""
    predictions = []
    references = []

    start = time.time()

    for example in tqdm(test_data_subset, desc=f"Evaluating {model_desc}"):
        inputs = tokenizer(
            example['input_text'],
            return_tensors="pt",
            max_length=max_input_length,
            truncation=True
        ).to(device)

        with torch.no_grad():
            outputs = model_eval.generate(
                **inputs,
                max_length=max_target_length,
                num_beams=4
            )

        pred = tokenizer.decode(outputs[0], skip_special_tokens=True)
        predictions.append(pred.strip())
        references.append(example['target_text'].strip())

    eval_time = time.time() - start

    # Compute metrics
    bleu = bleu_metric.compute(predictions=predictions, references=[[r] for r in references])
    rouge = rouge_metric.compute(predictions=predictions, references=references)
    em = sum([p == r for p, r in zip(predictions, references)]) / len(predictions) * 100
    bert = bertscore_metric.compute(
        predictions=predictions,
        references=references,
        lang="en",
        model_type="distilbert-base-uncased"
    )

    return {
        "model": model_desc,
        "bleu": bleu["score"],
        "rouge_l": rouge["rougeL"],
        "exact_match": em,
        "bertscore_f1": np.mean(bert["f1"]) * 100,
        "inference_time": eval_time,
        "latency_ms": eval_time / len(predictions) * 1000
    }

# Evaluate on first 100 samples for speed
test_subset = test_data[:100]
zeroshot_results = evaluate_baseline(base_model_zeroshot, test_subset, "Zero-shot FLAN-T5-base")

print(f"\nZero-shot Results (100 test samples):")
print(f"  BLEU-4: {zeroshot_results['bleu']:.2f}")
print(f"  ROUGE-L: {zeroshot_results['rouge_l']:.4f}")
print(f"  Exact Match: {zeroshot_results['exact_match']:.2f}%")
print(f"  BERTScore F1: {zeroshot_results['bertscore_f1']:.2f}")
print(f"  Latency per example: {zeroshot_results['latency_ms']:.2f}ms")
```

```python
# 13. FEW-SHOT EVALUATION
# ============================================================================
print("\n" + "="*80)
print("BENCHMARK: FEW-SHOT FLAN-T5-BASE (5-shot)")
print("="*80)

# Create few-shot prompt with 5 examples
few_shot_examples = train_data[:5]
few_shot_prompt = "\n\n".join([
    f"Instruction: {ex['input_text']}\nResponse: {ex['target_text']}"
    for ex in few_shot_examples
])

def evaluate_fewshot(model_eval, test_data_subset, few_shot_context):
    """Evaluate with few-shot prompting"""
    predictions = []
    references = []

    start = time.time()

    for example in tqdm(test_data_subset, desc="Few-shot evaluation"):
        # Prepend few-shot examples
        full_prompt = few_shot_context + f"\n\nInstruction: {example['input_text']}\nResponse:"

        inputs = tokenizer(
            full_prompt,
            return_tensors="pt",
            max_length=1024,  # Longer for few-shot context
            truncation=True
        ).to(device)

        with torch.no_grad():
            outputs = model_eval.generate(
                **inputs,
                max_length=max_target_length,
                num_beams=4
            )

        pred = tokenizer.decode(outputs[0], skip_special_tokens=True)
        predictions.append(pred.strip())
        references.append(example['target_text'].strip())

    eval_time = time.time() - start

    # Compute metrics
    bleu = bleu_metric.compute(predictions=predictions, references=[[r] for r in references])
    rouge = rouge_metric.compute(predictions=predictions, references=references)
    em = sum([p == r for p, r in zip(predictions, references)]) / len(predictions) * 100
    bert = bertscore_metric.compute(
        predictions=predictions,
        references=references,
        lang="en",
        model_type="distilbert-base-uncased"
    )

    return {
        "model": "Few-shot FLAN-T5-base (5-shot)",
        "bleu": bleu["score"],
        "rouge_l": rouge["rougeL"],
        "exact_match": em,
        "bertscore_f1": np.mean(bert["f1"]) * 100,
        "inference_time": eval_time,
        "latency_ms": eval_time / len(predictions) * 1000
    }

fewshot_results = evaluate_fewshot(base_model_zeroshot, test_subset, few_shot_prompt)

print(f"\nFew-shot Results (100 test samples):")
print(f"  BLEU-4: {fewshot_results['bleu']:.2f}")
print(f"  ROUGE-L: {fewshot_results['rouge_l']:.4f}")
print(f"  Exact Match: {fewshot_results['exact_match']:.2f}%")
print(f"  BERTScore F1: {fewshot_results['bertscore_f1']:.2f}")
print(f"  Latency per example: {fewshot_results['latency_ms']:.2f}ms")
```

```python
# 14. FINAL RESULTS COMPARISON
# ============================================================================
print("\n" + "="*80)
print("FINAL BENCHMARK COMPARISON")
print("="*80)

results_df = pd.DataFrame([
    {
        "Model": "Zero-shot FLAN-T5-base",
        "BLEU-4": f"{zeroshot_results['bleu']:.2f}",
        "ROUGE-L": f"{zeroshot_results['rouge_l']:.4f}",
        "Exact Match (%)": f"{zeroshot_results['exact_match']:.2f}",
        "BERTScore F1": f"{zeroshot_results['bertscore_f1']:.2f}",
        "Latency (ms)": f"{zeroshot_results['latency_ms']:.2f}"
    },
    {
        "Model": "Few-shot FLAN-T5-base (5-shot)",
        "BLEU-4": f"{fewshot_results['bleu']:.2f}",
        "ROUGE-L": f"{fewshot_results['rouge_l']:.4f}",
        "Exact Match (%)": f"{fewshot_results['exact_match']:.2f}",
        "BERTScore F1": f"{fewshot_results['bertscore_f1']:.2f}",
        "Latency (ms)": f"{fewshot_results['latency_ms']:.2f}"
    },
    {
        "Model": "FLAN-T5-base + LoRA (Ours)",
        "BLEU-4": f"{bleu_result['score']:.2f}",
        "ROUGE-L": f"{rouge_result['rougeL']:.4f}",
        "Exact Match (%)": f"{exact_match:.2f}",
        "BERTScore F1": f"{np.mean(bertscore_result['f1'])*100:.2f}",
        "Latency (ms)": f"{inference_time/len(all_predictions)*1000:.2f}"
    }
])

print("\n" + results_df.to_string(index=False))

# Save results
results_df.to_csv("alpaca_benchmark_results.csv", index=False)
print("\nResults saved to 'alpaca_benchmark_results.csv'")
```

```python
# 16. PLOT TRAINING CURVES
# ============================================================================
print("\n" + "="*80)
print("SAVING TRAINING CURVES")
print("="*80)

plt.figure(figsize=(10, 5))
epochs_range = range(1, num_epochs + 1)

# Plot Training Loss
if training_losses: # Ensure there's data to plot
    plt.plot(epochs_range[:len(training_losses)], training_losses, 'b-', label='Training Loss', marker='o')

# Plot Validation Loss, adjusting range if lengths differ
if validation_losses: # Ensure there's data to plot
    # If validation_losses has fewer points than training_losses,
    # assume it corresponds to the later epochs for plotting purposes.
    if len(validation_losses) == len(training_losses):
        plt.plot(epochs_range, validation_losses, 'r-', label='Validation Loss', marker='s')
    elif len(validation_losses) > 0 and len(validation_losses) < len(training_losses):
        # Assuming the available validation losses correspond to the end of training.
        val_epochs_range = list(epochs_range)[len(training_losses) - len(validation_losses):]
        plt.plot(val_epochs_range, validation_losses, 'r-', label='Validation Loss (partial)', marker='s')
    elif len(validation_losses) > len(training_losses):
        # This case is less likely but handled for robustness
        plt.plot(list(epochs_range)[:len(validation_losses)], validation_losses, 'r-', label='Validation Loss (truncated)', marker='s')

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Alpaca Fine-tuning: Training and Validation Loss')
plt.legend()
plt.grid(True)
plt.savefig('alpaca_training_curves.png', dpi=300, bbox_inches='tight')
print("Training curves saved to: alpaca_training_curves.png")
```
